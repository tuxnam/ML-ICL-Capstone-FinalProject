## Clean the data

### Remove Infinites
We have four columns with infinite values for approximately 10k rows (12%!). This is a considerable amount of data from the dataset. We decided to replace the infinite values with the median value of the column, to avoid losing information for the input of the model, but also avoid to create bias by zeroing the element. 

```python
# Select numeric columns
num_df = dataset.select_dtypes(include=[np.number])
# Check for infinite values
mask = ~num_df.applymap(np.isfinite)
# Print column names with infinite values
print(num_df.columns[mask.any()])

# Instead of dropping the rows with infinite value, as we have limited domain expertise on the sensors these values represent and 
# other useful inputs these rows could bring to the model, we decided to replace the infinite values with the median value for this
# column.

# 4 columns have infinite values: R1-PA:Z, R2-PA:Z, R3-PA:Z, R4-PA:Z

# Calculate the median for column 'R1-PA:Z' (excluding infinite values)
median_r1 = dataset['R1-PA:Z'].replace([np.inf, -np.inf], np.nan).median()
# Replace infinite values in column 'R1-PA:Z' with the median
dataset['R1-PA:Z'] = dataset['R1-PA:Z'].replace([np.inf, -np.inf], median_r1)

# Calculate the median for column 'R2-PA:Z' (excluding infinite values)
median_r2 = dataset['R2-PA:Z'].replace([np.inf, -np.inf], np.nan).median()
# Replace infinite values in column 'R2-PA:Z' with the median
dataset['R2-PA:Z'] = dataset['R2-PA:Z'].replace([np.inf, -np.inf], median_r2)

# Calculate the median for column 'R3-PA:Z' (excluding infinite values)
median_r3 = dataset['R3-PA:Z'].replace([np.inf, -np.inf], np.nan).median()
# Replace infinite values in column 'R3-PA:Z' with the median
dataset['R3-PA:Z'] = dataset['R3-PA:Z'].replace([np.inf, -np.inf], median_r3)

# Calculate the median for column 'R4-PA:Z' (excluding infinite values)
median_r4 = dataset['R4-PA:Z'].replace([np.inf, -np.inf], np.nan).median()
# Replace infinite values in column 'R4-PA:Z' with the median
dataset['R4-PA:Z'] = dataset['R4-PA:Z'].replace([np.inf, -np.inf], median_r4)

# Check if all infinites have been replaced
is_infinite = dataset.isin([np.inf, -np.inf]).sum()
print("Number of Infinites: ", is_infinite.sum())
```

### Scale the data
As we can observe in the analysis of the data, the data is not normalized and some columns have values above 100.000, while others are narrowed between 0 and 100, with negative values. 
There is a risk of feature dominance and introducing bias in the model.
We therefore need to scale the data.

Scaling options: min-max normalization and standardization.
Min-max is known to be useful for algorithms sensitive to feature scales (e.g., k-nearest neighbors), while standardization cam be used more in case of regression techniques or neural networks. 
We will therefore go with min-max normalization.

```python
from sklearn.preprocessing import MinMaxScaler

# Create a DataFrame with only numerical columns
num_df = dataset.select_dtypes(include=[np.number])

# Apply Min-Max Scaling to numerical columns
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(num_df))

# Combine scaled numerical columns with non-numerical columns
df_final = pd.concat([df_scaled, dataset['marker']])

# describe the final dataframe
df_final.describe()
```
